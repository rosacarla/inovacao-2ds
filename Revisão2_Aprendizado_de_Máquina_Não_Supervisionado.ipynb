### **Aprendizado Não Supervisionado (Grupos de Clientes)**

<p align="justify">  
<b>Contexto</b>: Neste desafio trabalharemos com dados sem rótulos ou "gabaritos". O objetivo é fazer a máquina agrupar clientes por comportamento usando o dataset <i>Mall Customers</i>, que contém informações de frequentadores de um shopping. Utilizaremos as colunas de Renda Anual (<i>Annual Income</i>) e Pontuação de Gastos (<i>Spending Score</i>) para identificar padrões ocultos e, assim, a própria IA define os grupos baseando-se apenas na proximidade dos dados entre os consumidores. </p>


👇 <b>Copie no código a seguir o link do dataset</b>: https://raw.githubusercontent.com/sharmaroshan/Clustering-of-Mall-Customers/master/Mall_Customers.csv

```python
# CLIQUE NO BOTÃO PLAY PARA EXECUTAR A CÉLULA
# Importa bibliotecas necessárias do Python
import pandas as pd # abre e organiza a tabela de dados dos clientes
from sklearn.cluster import KMeans # KMeans é o algoritmo de IA que criará grupos (clusters) automaticamente
import matplotlib.pyplot as plt # gera gráfico e visualiza os grupos na tela

# Preparação dos dados
url = 'https://raw.githubusercontent.com/sharmaroshan/Clustering-of-Mall-Customers/master/Mall_Customers.csv'

'COPIE-O-LINK-DO-DATASET!'

# Salva o dataset na variável df
df = pd.read_csv(url)

# Exibe o cabeçalho da tabela (5 primeiras linhas)
df.head()

# OBSERVE OS NOMES DAS COLUNAS DA TABELA COMPLETA
```

```python
# Seleção de dados para o treinamento do algoritmo KMeans
# Seleciona as colunas de Renda e Pontuação de Gasto para a variável X
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# Mostra cabeçalho da tabela reduzida pela seleção de dados armazenada em X
X.head()
```

```python
# --- DESAFIO 2 ---
# 1. Instancie (prepare) o algoritmo KMeans com n_clusters=5
# 2. Use o método fit_predict(X) para criar os grupos 'labels' (identificação numérica de grupos descobertos)

# PARA ATENDER AOS 2 ITENS DO DESAFIO 2, ESCREVA O SEU CÓDIGO AQUI:

# 1. Instanciando o modelo KMeans para encontrar 5 grupos distintos
kmeans = KMeans(n_clusters=5, random_state=42)

# 2. Treinando o modelo e prevendo os grupos de cada cliente ao mesmo tempo
labels = kmeans.fit_predict(X)

# Cria gráfico de dispersão para mostrar clusters ou agrupamentos de clientes
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=labels, cmap='viridis')

# Adiciona legendas dos eixos do gráfico
plt.xlabel("Renda Anual (k$)")       # Legenda do eixo X
plt.ylabel("Pontuação de Gastos (1-100)") # Legenda do eixo Y

plt.title("Grupos de Clientes Identificados")
plt.show()

# CLIQUE NO BOTÃO PLAY PARA EXECUTAR A CÉLULA E VER O GRÁFICO
```
