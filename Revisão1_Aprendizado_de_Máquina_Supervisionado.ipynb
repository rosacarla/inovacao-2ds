###  **Aprendizado Supervisionado (Filtro de Spam)**

<p align="justify">   
<b>Contexto</b>: Vamos ensinar a mÃ¡quina a ler mensagens e decidir se sÃ£o <b>SPAM</b> (perigosas ou indesejadas) ou <b>HAM</b> (seguras ou boas/comuns). Para isso, usaremos dados rotulados no dataset (conjunto de dados) <i>SMS Spam Collection</i>, que Ã© um banco de dados com 5.574 mensagens em inglÃªs rotuladas como <i>"ham"</i> (legÃ­timas) ou <i>"spam"</i> (indesejadas). O arquivo organiza uma mensagem por linha, dividida entre o rÃ³tulo de classificaÃ§Ã£o (v1) e o texto bruto da mensagem (v2). </p>  

ð <b>Copie no cÃ³digo a seguir o link do dataset</b>: https://raw.githubusercontent.com/mohitgupta-1O1/Kaggle-SMS-Spam-Collection-Dataset-/master/spam.csv






```python
# Importa bibliotecas necessÃ¡rias do Python
import pandas as pd # carrega e manipula tabelas de dados (DataFrames)
from sklearn.feature_extraction.text import CountVectorizer # transforma texto das mensagens em nÃºmeros (matriz de palavras)
from sklearn.naive_bayes import MultinomialNB # algoritmo de IA (Naive Bayes) que aprenderÃ¡ a classificar o Spam

# COMPLETE O CÃDIGO DE PREPARAÃÃO DOS DADOS
url = 'https://raw.githubusercontent.com/mohitgupta-1O1/Kaggle-SMS-Spam-Collection-Dataset-/master/spam.csv'
#'COPIE-O-LINK-DO-DATASET!'

# Salva o dataset na variÃ¡vel df
df = pd.read_csv(url, encoding='latin-1')[['v1', 'v2']]
# Define os nomes das 2 colunas
df.columns = ['label', 'text']

# Mostra cabeÃ§alho com 5 primeiras linhas do dataset
df.head()

# CLIQUE NO BOTÃO PLAY PARA EXECUTAR A CÃLULA
```

```python
# --- DESAFIO 1 ---
# 1. Instancie (prepare) o CountVectorizer na variÃ¡vel 'vetorizador' para que possamos converter os textos em nÃºmeros.
# 2. Use fit_transform no df['text'] para criar a matriz 'X'
# 3. Instancie o modelo MultinomialNB() e use o mÃ©todo .fit(X, df['label']) para que o algoritmo aprenda a relaÃ§Ã£o entre as palavras e as categorias.

# PARA ATENDER AOS 3 ITENS DO DESAFIO 1, ESCREVA O SEU CÃDIGO ABAIXO DE CADA COMENTÃRIO:

# 1. Instanciando o vetorizador
vetorizador = CountVectorizer()

# 2. Convertendo os textos em uma matriz numÃ©rica (X)
X = vetorizador.fit_transform(df['text'])

# 3. Criando o modelo e treinando (fit) com a matriz X e os rÃ³tulos de destino (label)
modelo = MultinomialNB()
modelo.fit(X, df['label'])

# SaÃ­da esperada: MultinomialNB(). Se aparecer este nome, seu classificador estÃ¡ pronto para testar novas mensagens.

# CLIQUE NO BOTÃO PLAY PARA EXECUTAR A CÃLULA
```

```python
# CLIQUE NO BOTÃO PLAY PARA EXECUTAR A CÃLULA

# Frases para teste que combina com o treino em InglÃªs
msg = ["You won a free prize! Click here now!"]
# msg = ["I'm a software development student from Brazil."]
# msg = ["Get a free software now, click here!"]

# Frases para teste que nÃ£o combinam com o treino em InglÃªs
# msg = ["Eu sou um estudante de programaÃ§Ã£o do Brazil."]
# msg = ["VocÃª ganhou um brinde! Clique aqui agora!"]
# msg = ["You tem um software free! Clicar no link!"]

# Teste da frase fornecida
pred = modelo.predict(vetorizador.transform(msg))

# Mostra a classificaÃ§Ã£o da mensagem
print(f"A mensagem foi classificada como: {pred}")

# RefaÃ§a o teste: coloque # na frase testada e retire # da frase seguinte.
# Mantenha o alinhamento Ã  esquerda da linha de comando ativada (sem #).
```
